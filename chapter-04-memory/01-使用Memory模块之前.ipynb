```python
import os
import dotenv
from langchain_community.chat_models import ChatTongyi

dotenv.load_dotenv()

chat_model = ChatTongyi(
    model="qwen3-max",
    api_key=os.getenv("DASHSCOPE_API_KEY")
)
```

```python
from langchain_core.messages import AIMessage, HumanMessage
from langchain_core.prompts import ChatPromptTemplate

def chat_with_model(answer):
    prompt_template = ChatPromptTemplate.from_messages([
        ("system", "你是一个人工智能助手。"),
        ("user", "{question}"),
    ])

    while True:
        chain = prompt_template | chat_model
        response = chain.invoke({"question": answer})

        #输出大模型的响应
        print(f"AI助手回答: {response.content}")

        #继续获取用户的问题
        user_input = input("你还有什么问题吗？(输入'退出'结束对话): ")

        if user_input == '退出':
            print("对话结束。")
            break

        #将上述新生成的消息存放在提示词模板的消息列表中
        prompt_template.messages.append(AIMessage(content=response.content))
        prompt_template.messages.append(HumanMessage(content=user_input))

chat_with_model("你好，介绍一下你自己。")
```
