```python
from langchain_community.chat_models import ChatTongyi
import os
import dotenv
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool
from langchain_community.tools.tavily_search import TavilySearchResults
from langchain_core.tools import StructuredTool
from langchain_experimental.utilities.python import PythonREPL

dotenv.load_dotenv()

os.environ["TAVILY_API_KEY"] = os.getenv("TAVILY_API_KEY")
search = TavilySearchResults(max_results=3)

python_repl = PythonREPL()

calc_tool = Tool(
    name="Calculator",
    func=python_repl.run,
    description="ä¸€ä¸ªå¯ä»¥è¿›è¡Œæ•°å­¦è®¡ç®—çš„å·¥å…·ï¼Œä¾‹å¦‚è®¡ç®—ç™¾åˆ†æ¯”å˜åŒ–",
)

# search_tool = StructuredTool.from_function(
#     func=search.run,
#     name="Search",
#     description="ç”¨äºæ£€ç´¢äº’è”ç½‘ä¸Šçš„ä¿¡æ¯",
# )

# search_tool = Tool(
#     name="Search",
#     func=search.run,
#     description="ç”¨äºæ£€ç´¢äº’è”ç½‘ä¸Šçš„ä¿¡æ¯",
# )


os.environ["DASHSCOPE_API_KEY"] = os.getenv("DASHSCOPE_API_KEY")
model = ChatTongyi(
    model="qwen3-max",
)

agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION

agent_executor = initialize_agent(
    tools=[search, calc_tool],
    llm=model,
    agent=agent,
    verbose=True,
)

result = agent_executor.invoke("è‹¹æœå½“å‰çš„è‚¡ä»·æ˜¯å¤šå°‘ï¼Ÿæ¯”å»å¹´ä¸Šæ¶¨äº†ç™¾åˆ†ä¹‹å‡ ï¼Ÿ")
print(result)
```
